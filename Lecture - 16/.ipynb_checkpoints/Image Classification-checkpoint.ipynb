# Create Validation Data

```python
import os, shutil
```

```python
if not os.path.isdir("val_images"):
    os.mkdir("val_images")
    
clases = ["cats", "dogs", "horses", "humans"]
```

```python
for c in clases:
    p = os.path.join("val_images/" , c)
    if not os.path.isdir(p):
        os.mkdir(p)
```

```python
SPLIT = 0.9


for dirc in os.listdir("images/"):
    p = "images/" + dirc
    imgs = os.listdir(p)
    split_size = int(SPLIT *len(imgs))
    files_to_move = imgs[split_size:]
    
    for img_f in files_to_move:
        src = os.path.join(p,img_f)
        des = os.path.join("val_images/"+dirc)
        shutil.move(src,des)
#         print(src, des, sep="\n")
```

```python
for dirc in os.listdir("images/"):
    p = "images/" + dirc
    imgs = os.listdir(p)
    print(p , len(imgs))
print()
for dirc in os.listdir("val_images/"):
    p = "val_images/" + dirc
    imgs = os.listdir(p)
    print(p , len(imgs))
```

## Create Model

```python
from keras.layers import *
from keras.models import Sequential
```

```python
model = Sequential()

model.add(Conv2D(32, (3,3), activation="relu", input_shape = (100,100,3)))
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.3))

model.add(Conv2D(64, (3,3), activation="relu",))
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.3))

model.add(Conv2D(128, (3,3), activation="relu",))
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.3))

model.add(Flatten())
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(4, activation="softmax"))
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
train_data = image.ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        vertical_flip=True,
        width_shift_range=0.2,
        horizontal_flip=True)
val_data = image.ImageDataGenerator(rescale=1.0/255.)
```

```python
train_generator  = train_data.flow_from_directory("images/",
                                                  target_size=(100,100),
                                                 batch_size=32,
                                                class_mode="categorical",
                                                 shuffle=True)
```

```python
val_generator  = train_data.flow_from_directory("val_images/",
                                                  target_size=(100,100),
                                                 batch_size=32,
                                                class_mode="categorical",
                                                 shuffle=True)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint = ModelCheckpoint("model", monitor="val_acc", save_best_only=True)
earlystop = EarlyStopping(monitor="val_acc", min_delta=0.05, patience=8, restore_best_weights=True)
```

```python
hist = model.fit_generator(train_generator, steps_per_epoch=25,epochs=30,
                           validation_data=val_generator,
                           validation_steps=3,
                           callbacks=[checkpoint,earlystop])
```

```python
plt.plot(hist.history['acc'])
plt.plot(hist.history['val_acc'])
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
model.evaluate_generator(val_generator, steps=3)
```
