```python
import matplotlib.pyplot as plt
from keras.preprocessing import image
import numpy as np
```

```python
img =image.load_img("horse-12.jpg", target_size=(100,100))
img = image.img_to_array(img)
img = np.expand_dims(img, axis=0)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
train_datagen = image.ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        vertical_flip=True,
        width_shift_range=0.2,
        horizontal_flip=True)
```

```python
Train = train_datagen.flow(img, batch_size=1)
```

```python
i =0
for batch in Train:
    plt.imshow(image.img_to_array(batch[0]))
    plt.show()
    i+=1
    if i%5==0:
        break
    
```
